```python
"""
Page Flip Detection System

This module implements a machine learning system to detect page flips in video frames.
It uses motion features and image analysis through a CNN model to classify frames as
either containing a page flip or not.

Example usage:
    # Create dataset and analyze
    df = create_dataset_df(base_path)
    analyze_dataset(df)

    # Add motion features and preprocess
    df = calculate_optimized_motion_features(df)
    visualize_preprocessing(df)

    # Train model
    model, history = train_model(...)

    # Evaluate
    optimal_threshold = find_optimal_threshold(model, val_loader)
    test_metrics = test_with_threshold(model, test_loader, threshold=optimal_threshold)
"""


from google.colab import drive
drive.mount('/content/drive')

import os
import time
import pandas as pd
import numpy as np
from PIL import Image, ImageEnhance, ImageFilter
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score, recall_score
import seaborn as sns
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
import cv2
from pathlib import Path
import random
import multiprocessing
from functools import partial


# Set random seed for reproducibility
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)
torch.set_num_threads(min(4, os.cpu_count()))

# Hyperparameters for optimization
IMAGE_SIZE = 96  # Reduced from 128 to 96 for faster processing
BATCH_SIZE = 128  # Increased batch size
MOTION_DOWNSCALE = 64  # Size to downscale images for motion calculation
NUM_EPOCHS = 10
EARLY_STOP_PATIENCE = 3
WORKERS = min(4, os.cpu_count())
MOTION_CACHE_FILE = 'motion_features_cache.npz'  # Cache file for motion features
USE_MOTION_FEATURES = True
USE_REGULARIZATION = True  # Add regularization to help with validation stability
PREVENT_OVERFITTING = True  # Use techniques to prevent overfitting

# Set device (GPU if available, otherwise CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Set the base path
base_path = "/content/drive/MyDrive/images"

def create_dataset_df(base_path):
    """
    Create a DataFrame containing image paths and labels with additional metadata.

    This function scans the directory structure looking for page flip images, extracts
    metadata from the filenames, and builds a comprehensive DataFrame with all the information.

    Parameters:
    -----------
    base_path : str
        Path to the root directory containing the 'images' folder with training and testing data

    Returns:
    --------
    pandas.DataFrame
        DataFrame with columns: 'image_path', 'label', 'video_id', 'frame_number',
        'split' (training/testing), 'sequence_position'

    Example:
    --------
    >>> base_path = "/content/drive/MyDrive/images"
    >>> df = create_dataset_df(base_path)
    >>> print(f"Found {len(df)} images")
    Found 5240 images
    """

    data = []
    images_path = os.path.join(base_path, 'images')

    print(f"Checking directory structure:")
    print(f"Base path: {base_path}")
    print(f"Images path: {images_path}")

    for split in ['training', 'testing']:
        split_path = os.path.join(images_path, split)
        print(f"\nProcessing {split} data...")

        for label in ['flip', 'notflip']:
            label_path = os.path.join(split_path, label)

            if not os.path.exists(label_path):
                print(f"Warning: Path does not exist: {label_path}")
                continue

            files = [f for f in os.listdir(label_path) if f.endswith('.jpg')]
            print(f"{split}/{label}: {len(files)} images found")

            for img_name in files:
                try:
                    parts = img_name.split('_')
                    video_id = parts[0]
                    frame_number = int(parts[1].split('.')[0])

                    # Calculate the sequence info based on frame numbers
                    data.append({
                        'image_path': os.path.join(label_path, img_name),
                        'label': 1 if label == 'flip' else 0,
                        'video_id': video_id,
                        'frame_number': frame_number,
                        'split': split,
                        'sequence_position': get_sequence_position(frame_number)
                    })
                except Exception as e:
                    print(f"Error processing {img_name}: {str(e)}")

    return pd.DataFrame(data)

def get_sequence_position(frame_number):
    """
    Categorize frames into beginning, middle, or end of a sequence based on frame number.

    Parameters:
    -----------
    frame_number : int
        The frame number within a video sequence

    Returns:
    --------
    str
        One of: "beginning", "middle", or "end"

    Example:
    --------
    >>> get_sequence_position(5)
    'beginning'
    >>> get_sequence_position(15)
    'middle'
    >>> get_sequence_position(25)
    'end'
    """

    if frame_number < 10:
        return "beginning"
    elif frame_number < 20:
        return "middle"
    else:
        return "end"

def analyze_dataset(df):
    """
    Print comprehensive dataset statistics with enhanced analysis.

    This function analyzes the dataset distribution by split (train/test),
    label (flip/notflip), and provides visualizations of class balance and
    sequence position distributions.

    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the dataset information (from create_dataset_df)

    Returns:
    --------
    None
        Prints statistics and displays visualizations

    Example:
    --------
    >>> df = create_dataset_df("/content/drive/MyDrive/images")
    >>> analyze_dataset(df)

    Dataset Overview:
    --------------------------------------------------
    Total number of images: 5240

    Distribution by split:
    training    3928
    testing     1312
    ...
    """

    print("\n Dataset Overview:")
    print("-" * 50)
    print(f"Total number of images: {len(df)}")

    print("\n Distribution by split:")
    print(df['split'].value_counts())

    print("\n Distribution by label:")
    label_dist = df['label'].value_counts()
    print("0 (NotFlip):", label_dist.get(0, 0))
    print("1 (Flip):", label_dist.get(1, 0))

    print("\n Detailed distribution:")
    dist_matrix = pd.crosstab(df['split'], df['label'])
    dist_matrix.columns = ['NotFlip', 'Flip']
    print(dist_matrix)

    print("\n Unique videos in each split:")
    print(df.groupby('split')['video_id'].nunique())

    # Additional analysis of sequence positions
    print("\n Sequence position distribution:")
    print(pd.crosstab([df['split'], df['label']], df['sequence_position']))

    # Check for class imbalance by video
    print("\n Class distribution by video:")
    video_class_dist = df.groupby(['video_id', 'label']).size().unstack().fillna(0)
    imbalanced_videos = video_class_dist[video_class_dist[0] / (video_class_dist[1] + 1e-5) > 2]
    if not imbalanced_videos.empty:
        print(f"Found {len(imbalanced_videos)} videos with significant class imbalance")

    # Visualize a few examples of frame distributions
    plt.figure(figsize=(12, 6))
    sns.countplot(x='frame_number', hue='label', data=df[df['video_id'].isin(df['video_id'].unique()[:5])])
    plt.title('Frame Distribution by Label (Sample Videos)')
    plt.xlabel('Frame Number')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

def extract_optimized_motion_features(image_path, prev_image_path=None):
    """
    Extract motion-related features using simplified, faster approach.

    Calculates the motion between two consecutive frames using frame differencing
    and extracts basic statistics about the motion.

    Parameters:
    -----------
    image_path : str
        Path to the current image file
    prev_image_path : str, optional
        Path to the previous image file. If None, returns zeros

    Returns:
    --------
    numpy.ndarray
        Array of 3 motion features: mean_motion, std_motion, max_motion

    Example:
    --------
    >>> features = extract_optimized_motion_features(
    ...     'images/video1_10.jpg',
    ...     'images/video1_09.jpg'
    ... )
    >>> print(features)
    [12.5, 24.3, 87.0]
    """

    # If no previous image, return zeros
    if prev_image_path is None or not os.path.exists(prev_image_path):
        return np.zeros(3)

    # Read images directly as grayscale for faster processing
    curr_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    prev_img = cv2.imread(prev_image_path, cv2.IMREAD_GRAYSCALE)

    if curr_img is None or prev_img is None:
        return np.zeros(3)

    # Resize images to much smaller size for faster processing
    curr_small = cv2.resize(curr_img, (MOTION_DOWNSCALE, MOTION_DOWNSCALE))
    prev_small = cv2.resize(prev_img, (MOTION_DOWNSCALE, MOTION_DOWNSCALE))

    # Use simple frame differencing instead of optical flow
    diff = cv2.absdiff(curr_small, prev_small)

    # Extract basic motion statistics
    mean_motion = np.mean(diff)
    std_motion = np.std(diff)
    max_motion = np.max(diff)

    return np.array([mean_motion, std_motion, max_motion])

def process_video_motion(video_data):
    """
    Process motion features for a single video.

    Takes a video group (all frames from one video) and calculates motion
    features between consecutive frames.

    Parameters:
    -----------
    video_data : tuple
        Tuple containing (video_id, video_df) where video_df is a DataFrame
        containing all frames for a single video

    Returns:
    --------
    tuple
        (video_id, motion_features, indices) where motion_features is a numpy array
        of motion features and indices are the DataFrame indices

    Example:
    --------
    >>> video_df = df[df['video_id'] == 'video1']
    >>> video_id, features, indices = process_video_motion(('video1', video_df))
    >>> print(f"Processed {len(features)} frames for video {video_id}")
    Processed 30 frames for video video1
    """

    video_id, video_df = video_data
    sorted_df = video_df.sort_values('frame_number')
    motion_features = np.zeros((len(sorted_df), 3))

    prev_path = None
    for i, (_, row) in enumerate(sorted_df.iterrows()):
        features = extract_optimized_motion_features(row['image_path'], prev_path)
        motion_features[i] = features
        prev_path = row['image_path']

    return video_id, motion_features, sorted_df.index.tolist()

def calculate_optimized_motion_features(df, use_cache=True):
    """
    Calculate motion features with optimized parallel processing and caching.

    This function adds motion features to the dataset, processing each video independently
    and in parallel. Results are cached to disk for faster loading in future runs.

    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the dataset information
    use_cache : bool, default=True
        Whether to use cached features if available

    Returns:
    --------
    pandas.DataFrame
        Updated DataFrame with additional columns: 'mean_motion', 'std_motion', 'max_motion'

    Example:
    --------
    >>> df = create_dataset_df("/content/drive/MyDrive/images")
    >>> df = calculate_optimized_motion_features(df)
    >>> print(df[['mean_motion', 'std_motion', 'max_motion']].head())
       mean_motion  std_motion  max_motion
    0        12.31       23.45        89.0
    1        15.67       28.92        95.5
    ...
    """

    cache_path = os.path.join(os.path.dirname(df['image_path'].iloc[0]), MOTION_CACHE_FILE)

    # Check if cached features exist
    if use_cache and os.path.exists(cache_path):
        print(f"Loading cached motion features from {cache_path}")
        cached_data = np.load(cache_path)

        # Verify the cache matches our dataframe
        if len(cached_data['indices']) == len(df):
            print("Using cached motion features")
            for i, feature_name in enumerate(['mean_motion', 'std_motion', 'max_motion']):
                df[feature_name] = cached_data['features'][:, i]
            return df
        else:
            print("Cache size mismatch, recalculating features")

    print("\n🔄 Calculating motion features (optimized)...")
    start_time = time.time()

    # Group by video_id for parallel processing
    video_groups = [(video_id, group) for video_id, group in df.groupby('video_id')]

    # Use multiprocessing for faster computation
    with multiprocessing.Pool(processes=WORKERS) as pool:
        results = list(tqdm(
            pool.imap(process_video_motion, video_groups),
            total=len(df['video_id'].unique()),
            desc="Processing videos"
        ))

    # Collect results
    all_features = np.zeros((len(df), 3))
    indices = []

    for video_id, features, idx in results:
        for i, df_idx in enumerate(idx):
            df_pos = df.index.get_loc(df_idx)
            all_features[df_pos] = features[i]
            indices.append(df_pos)

    # Save to cache
    np.savez(cache_path, features=all_features, indices=np.array(indices))

    # Add to dataframe
    for i, feature_name in enumerate(['mean_motion', 'std_motion', 'max_motion']):
        df[feature_name] = all_features[:, i]

    print(f"Motion features calculated in {(time.time() - start_time)/60:.2f} minutes")

    # Visualize motion features
    plt.figure(figsize=(15, 10))
    feature_names = ['mean_motion', 'std_motion', 'max_motion']

    for i, feature in enumerate(feature_names):
        plt.subplot(1, 3, i+1)
        sns.histplot(x=feature, hue='label', data=df)
        plt.title(f'Distribution of {feature}')

    plt.tight_layout()
    plt.show()

    return df

def preprocess_image(image, target_size=(IMAGE_SIZE, IMAGE_SIZE), preprocessing_level='basic'):
    """
    Optimized image preprocessing with multiple levels of enhancement.

    Parameters:
    -----------
    image : PIL.Image
        Input image to preprocess
    target_size : tuple, default=(IMAGE_SIZE, IMAGE_SIZE)
        Output size as (width, height)
    preprocessing_level : str, default='basic'
        Level of preprocessing to apply:
        - 'none': Only resize
        - 'basic': Crop unnecessary background + resize
        - 'full': Crop + contrast enhancement + sharpening + resize

    Returns:
    --------
    PIL.Image
        Preprocessed image

    Example:
    --------
    >>> from PIL import Image
    >>> img = Image.open('images/video1_10.jpg')
    >>> # Basic preprocessing
    >>> basic_img = preprocess_image(img, preprocessing_level='basic')
    >>> # Full preprocessing with custom size
    >>> full_img = preprocess_image(img, target_size=(64, 64), preprocessing_level='full')
    """
    # Convert to RGB
    image = image.convert('RGB')

    # Apply cropping only if not 'none' level
    if preprocessing_level != 'none':
        # Crop unnecessary background
        bbox = image.getbbox()
        if bbox:
            padding = 5  # Reduced padding to focus more on content
            left, top, right, bottom = bbox
            left = max(0, left - padding)
            top = max(0, top - padding)
            right = min(image.width, right + padding)
            bottom = min(image.height, bottom + padding)
            image = image.crop((left, top, right, bottom))

    # Apply additional processing only for 'full' level
    if preprocessing_level == 'full':
        # Enhance contrast for better feature detection
        enhancer = ImageEnhance.Contrast(image)
        image = enhancer.enhance(1.2)

        # Enhance sharpness to make edges more visible
        enhancer = ImageEnhance.Sharpness(image)
        image = enhancer.enhance(1.1)

    # Resize
    image = image.resize(target_size)

    return image

def visualize_preprocessing(df, num_samples=2):
    """
    Visualize the impact of preprocessing on random images from the dataset.

    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing the dataset information
    num_samples : int, default=2
        Number of random samples to visualize

    Returns:
    --------
    None
        Displays visualization of original and preprocessed images

    Example:
    --------
    >>> df = create_dataset_df("/content/drive/MyDrive/images")
    >>> visualize_preprocessing(df, num_samples=3)

    Preprocessing visualization:
    [Displays 3 rows of images showing original, basic, and full preprocessing]
    """

    print("\n Preprocessing visualization:")

    plt.figure(figsize=(15, 4 * num_samples))
    sample_indices = np.random.choice(len(df), num_samples, replace=False)

    for i, idx in enumerate(sample_indices):
        row = df.iloc[idx]

        with Image.open(row['image_path']) as img:
            # Original image
            plt.subplot(num_samples, 3, i*3 + 1)
            plt.imshow(img)
            plt.title(f"Original - {os.path.basename(row['image_path'])}")
            plt.axis('off')

            # Basic preprocessing
            plt.subplot(num_samples, 3, i*3 + 2)
            preprocessed = preprocess_image(img, preprocessing_level='basic')
            plt.imshow(preprocessed)
            plt.title("Basic Preprocessing")
            plt.axis('off')

            # Full preprocessing
            plt.subplot(num_samples, 3, i*3 + 3)
            full_preprocessed = preprocess_image(img, preprocessing_level='full')
            plt.imshow(full_preprocessed)
            plt.title("Full Preprocessing")
            plt.axis('off')

    plt.tight_layout()
    plt.show()

class PageFlipDataset(Dataset):
    def __init__(self, df, transform=None, include_motion_features=True,
                 augment=False, preprocessing_level='basic', mixup=False):
        """
    Custom PyTorch Dataset for page flip detection.

    Handles loading images, preprocessing, and optional motion features and augmentation.

    Attributes:
    -----------
    df : pandas.DataFrame
        DataFrame containing image paths and labels
    transform : callable
        PyTorch transforms to apply to images
    include_motion_features : bool
        Whether to include motion features with each sample
    augment : bool
        Whether to apply data augmentation
    preprocessing_level : str
        Level of preprocessing to apply (none/basic/full)
    mixup : bool
        Whether to apply mixup augmentation

    Example:
    --------
    >>> train_df = df[df['split'] == 'training']
    >>> train_transform = transforms.Compose([
    ...     transforms.ToTensor(),
    ...     transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ... ])
    >>> train_dataset = PageFlipDataset(
    ...     train_df,
    ...     transform=train_transform,
    ...     include_motion_features=True,
    ...     augment=True
    ... )
    >>> image, label, motion_features = train_dataset[0]
    >>> print(f"Image shape: {image.shape}, Label: {label}, Motion features: {motion_features}")
    Image shape: torch.Size([3, 96, 96]), Label: 0.0, Motion features: tensor([12.5000, 24.3000, 87.0000])
    """


        self.df = df
        self.transform = transform
        self.include_motion_features = include_motion_features
        self.augment = augment
        self.preprocessing_level = preprocessing_level
        self.mixup = mixup

        # Store motion feature column names
        self.motion_feature_cols = [col for col in df.columns
                                   if col in ['mean_motion', 'std_motion', 'max_motion']]

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = row['image_path']
        label = row['label']

        with Image.open(img_path) as img:
            # Apply preprocessing
            img = preprocess_image(img, preprocessing_level=self.preprocessing_level)

            # Apply augmentation if enabled
            if self.augment and random.random() > 0.5:
                # Random rotation (small angles only)
                angle = random.uniform(-5, 5)
                img = img.rotate(angle, expand=False, resample=Image.BILINEAR)

                # Random brightness adjustment (subtle)
                enhancer = ImageEnhance.Brightness(img)
                img = enhancer.enhance(random.uniform(0.95, 1.05))

            if self.transform:
                img = self.transform(img)

        # Include motion features if requested
        if self.include_motion_features and len(self.motion_feature_cols) > 0:
            motion_features = torch.tensor([row[col] for col in self.motion_feature_cols],
                                          dtype=torch.float32)
            return img, torch.tensor(label, dtype=torch.float32), motion_features

        return img, torch.tensor(label, dtype=torch.float32)

class OptimizedPageFlipNet(nn.Module):
    def __init__(self, include_motion_features=True, kernel_sizes=None, dropout_rate=0.3):
        """
    Optimized CNN model for page flip detection.

    This neural network combines image features with optional motion features
    to predict whether an image contains a page flip.

    Parameters:
    -----------
    include_motion_features : bool, default=True
        Whether to include motion features in the model
    kernel_sizes : list, default=None
        List of kernel sizes for each conv layer
    dropout_rate : float, default=0.3
        Dropout rate for regularization

    Example:
    --------
    >>> # Create model with default parameters
    >>> model = OptimizedPageFlipNet(include_motion_features=True)
    >>>
    >>> # Forward pass with image only
    >>> image = torch.randn(1, 3, 96, 96)
    >>> output = model(image)
    >>> print(f"Output probability: {output.item():.4f}")
    Output probability: 0.3782
    >>>
    >>> # Forward pass with image and motion features
    >>> motion_features = torch.tensor([[12.5, 24.3, 87.0]], dtype=torch.float32)
    >>> output = model(image, motion_features)
    >>> print(f"Output probability: {output.item():.4f}")
    Output probability: 0.4215
    """
        super().__init__()

        # Set default kernel sizes if not provided
        if kernel_sizes is None:
            kernel_sizes = [3, 5, 3, 3]

        self.include_motion_features = include_motion_features
        motion_feature_size = 3 if include_motion_features else 0  # Reduced from 5 to 3

        # First Block - Extract basic features
        self.features_block1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=kernel_sizes[0], padding=kernel_sizes[0]//2),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Dropout2d(0.1) if PREVENT_OVERFITTING else nn.Identity(),  # Light dropout
        )

        # Second Block - Edge detection
        self.features_block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=kernel_sizes[1], padding=kernel_sizes[1]//2),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Dropout2d(0.15) if PREVENT_OVERFITTING else nn.Identity(),  # More dropout
        )

        # Third Block - Higher-level features
        self.features_block3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=kernel_sizes[2], padding=kernel_sizes[2]//2),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
        )

        # Reduced motion detection block size from 256 to 192 channels
        self.motion_detection = nn.Sequential(
            nn.Conv2d(128, 192, kernel_size=kernel_sizes[3], padding=kernel_sizes[3]//2),
            nn.BatchNorm2d(192),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1))
        )

        # Feature fusion layer (smaller)
        if include_motion_features:
            self.feature_fusion = nn.Sequential(
                nn.Linear(192 + motion_feature_size, 96),
                nn.ReLU(),
                nn.Dropout(0.2) if PREVENT_OVERFITTING else nn.Identity()
            )
            classifier_input_size = 96
        else:
            classifier_input_size = 192

        # Classification head
        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(classifier_input_size, 32),  # Reduced from 64 to 32
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, x, motion_features=None):
        # Image feature extraction
        x = self.features_block1(x)
        x = self.features_block2(x)
        x = self.features_block3(x)
        x = self.motion_detection(x)
        x = x.view(x.size(0), -1)  # Flatten

        # Combine with motion features if available
        if self.include_motion_features and motion_features is not None:
            x = torch.cat([x, motion_features], dim=1)
            x = self.feature_fusion(x)

        # Classification
        x = self.classifier(x)
        return x



class EarlyStopping:
    def __init__(self, patience=3, min_delta=0.001, restore_best_weights=True):

        """
    Early stopping handler to prevent overfitting during training.

    Monitors validation loss and stops training when no improvement is seen
    for a specified number of epochs.

    Parameters:
    -----------
    patience : int, default=3
        Number of epochs with no improvement after which training stops
    min_delta : float, default=0.001
        Minimum change in monitored value to qualify as improvement
    restore_best_weights : bool, default=True
        Whether to restore model to best weights when stopped

    Example:
    --------
    >>> early_stopping = EarlyStopping(patience=3)
    >>> for epoch in range(10):
    ...     # Train model
    ...     train_loss = ...
    ...     val_loss = ...
    ...
    ...     # Check early stopping
    ...     improved = early_stopping(val_loss, model)
    ...     if improved:
    ...         print(f"Epoch {epoch+1}: Model improved!")
    ...
    ...     if early_stopping.early_stop:
    ...         print(f"Early stopping triggered after epoch {epoch+1}")
    ...         break
    ...
    >>> # Restore best model weights
    >>> early_stopping.restore_model(model)
    """

        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = float('inf')
        self.early_stop = False
        self.restore_best_weights = restore_best_weights
        self.best_model_state = None

    def __call__(self, val_loss, model=None):
        """Returns True if model improved"""
        if val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
            if model is not None and self.restore_best_weights:
                self.best_model_state = model.state_dict().copy()
            return True
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
            return False

    def restore_model(self, model):
        """Restore model to best weights"""
        if self.best_model_state is not None and self.restore_best_weights:
            model.load_state_dict(self.best_model_state)
            return True
        return False

def get_transforms(use_augmentation=False):
    """
    Get transforms for training and validation datasets.

    Parameters:
    -----------
    use_augmentation : bool, default=False
        Whether to include data augmentation in the training transform

    Returns:
    --------
    tuple
        (train_transform, val_transform) - PyTorch transforms for training and validation

    Example:
    --------
    >>> train_transform, val_transform = get_transforms(use_augmentation=True)
    >>> # Use transforms to create datasets
    >>> train_dataset = PageFlipDataset(train_df, transform=train_transform)
    >>> val_dataset = PageFlipDataset(val_df, transform=val_transform)
    """

    # Basic normalization
    base_transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    if use_augmentation:
        # Subtle augmentations to improve generalization without drastic changes
        train_transform = transforms.Compose([
            transforms.ColorJitter(brightness=0.05, contrast=0.05),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])
        return train_transform, base_transform

    return base_transform, base_transform

def calculate_metrics(outputs, labels, threshold=0.5):
    """
    Calculate comprehensive classification metrics with adjustable threshold.

    Parameters:
    -----------
    outputs : torch.Tensor
        Model output probabilities
    labels : torch.Tensor
        Ground truth labels
    threshold : float, default=0.5
        Classification threshold

    Returns:
    --------
    dict
        Dictionary containing metrics:
        'accuracy', 'precision', 'recall', 'f1', 'specificity', 'balanced_acc',
        'predictions', 'true_labels', 'probabilities'

    Example:
    --------
    >>> # Evaluate model outputs
    >>> outputs = torch.tensor([0.7, 0.3, 0.9, 0.2, 0.6])
    >>> labels = torch.tensor([1.0, 0.0, 1.0, 0.0, 1.0])
    >>>
    >>> # Calculate metrics with default threshold (0.5)
    >>> metrics = calculate_metrics(outputs, labels)
    >>> print(f"F1 Score: {metrics['f1']:.4f}, Accuracy: {metrics['accuracy']:.4f}")
    F1 Score: 0.8571, Accuracy: 0.8000
    >>>
    >>> # Calculate metrics with custom threshold (0.7)
    >>> metrics_custom = calculate_metrics(outputs, labels, threshold=0.7)
    >>> print(f"F1 Score: {metrics_custom['f1']:.4f}, Accuracy: {metrics_custom['accuracy']:.4f}")
    F1 Score: 0.6667, Accuracy: 0.6000
    """

    predictions = (outputs > threshold).float()

    # Basic counts
    tp = ((predictions == 1) & (labels == 1)).sum().float()
    fp = ((predictions == 1) & (labels == 0)).sum().float()
    tn = ((predictions == 0) & (labels == 0)).sum().float()
    fn = ((predictions == 0) & (labels == 1)).sum().float()

    # Metrics calculations
    accuracy = (tp + tn) / (tp + fp + tn + fn + 1e-7)
    precision = tp / (tp + fp + 1e-7)
    recall = tp / (tp + fn + 1e-7)
    f1 = 2 * (precision * recall) / (precision + recall + 1e-7)
    specificity = tn / (tn + fp + 1e-7)

    # Balanced accuracy
    balanced_acc = (recall + specificity) / 2

    return {
        'accuracy': accuracy.item(),
        'precision': precision.item(),
        'recall': recall.item(),
        'f1': f1.item(),
        'specificity': specificity.item(),
        'balanced_acc': balanced_acc.item(),
        'predictions': predictions.cpu().numpy(),
        'true_labels': labels.cpu().numpy(),
        'probabilities': outputs.detach().cpu().numpy()
    }

def train_epoch(model, train_loader, criterion, optimizer, include_motion_features=True,
                l2_reg=0.0001, clip_grad_norm=1.0):
    """
    Train model for one epoch with regularization to prevent overfitting.

    Parameters:
    -----------
    model : torch.nn.Module
        Model to train
    train_loader : torch.utils.data.DataLoader
        DataLoader for training data
    criterion : torch.nn.Module
        Loss function
    optimizer : torch.optim.Optimizer
        Optimizer
    include_motion_features : bool, default=True
        Whether batches include motion features
    l2_reg : float, default=0.0001
        L2 regularization strength
    clip_grad_norm : float, default=1.0
        Gradient clipping norm

    Returns:
    --------
    dict
        Dictionary containing training metrics and loss

    Example:
    --------
    >>> # Setup model, criterion, optimizer
    >>> model = OptimizedPageFlipNet(include_motion_features=True)
    >>> criterion = nn.BCELoss()
    >>> optimizer = optim.Adam(model.parameters(), lr=0.001)
    >>>
    >>> # Train for one epoch
    >>> train_metrics = train_epoch(model, train_loader, criterion, optimizer)
    >>> print(f"Training loss: {train_metrics['loss']:.4f}, F1: {train_metrics['f1']:.4f}")
    Training loss: 0.3824, F1: 0.8567
    """

    model.train()
    running_loss = 0.0
    all_outputs = []
    all_labels = []

    for batch in tqdm(train_loader, desc='Training'):
        # Handle batch structure based on whether motion features are included
        if include_motion_features and len(batch) == 3:
            inputs, labels, motion_features = batch
            inputs = inputs.to(device)
            labels = labels.to(device)
            motion_features = motion_features.to(device)
            outputs = model(inputs, motion_features)
        else:
            inputs, labels = batch
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)

        # Calculate loss with L2 regularization
        loss = criterion(outputs.squeeze(), labels)

        # Add L2 regularization if enabled
        if USE_REGULARIZATION and l2_reg > 0:
            l2_loss = 0
            for param in model.parameters():
                l2_loss += torch.norm(param, 2)
            loss += l2_reg * l2_loss

        # Update weights
        optimizer.zero_grad(set_to_none=True)
        loss.backward()

        # Gradient clipping to prevent exploding gradients
        if PREVENT_OVERFITTING and clip_grad_norm > 0:
            torch.nn.utils.clip_grad_norm_(model.parameters(), clip_grad_norm)

        optimizer.step()

        running_loss += loss.item()
        all_outputs.extend(outputs.squeeze().detach().cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    metrics = calculate_metrics(torch.tensor(all_outputs), torch.tensor(all_labels))
    metrics['loss'] = running_loss / len(train_loader)

    return metrics

def validate_epoch(model, val_loader, criterion, include_motion_features=True, threshold=0.5):
    """
    Validate model performance with adjustable threshold.

    Parameters:
    -----------
    model : torch.nn.Module
        Model to validate
    val_loader : torch.utils.data.DataLoader
        DataLoader for validation data
    criterion : torch.nn.Module
        Loss function
    include_motion_features : bool, default=True
        Whether batches include motion features
    threshold : float, default=0.5
        Classification threshold

    Returns:
    --------
    dict
        Dictionary containing validation metrics and loss

    Example:
    --------
    >>> # Validate model
    >>> val_metrics = validate_epoch(model, val_loader, criterion)
    >>> print(f"Validation loss: {val_metrics['loss']:.4f}, F1: {val_metrics['f1']:.4f}")
    Validation loss: 0.4215, F1: 0.8123
    >>>
    >>> # Validate with custom threshold
    >>> val_metrics_custom = validate_epoch(model, val_loader, criterion, threshold=0.7)
    >>> print(f"Validation loss: {val_metrics_custom['loss']:.4f}, F1: {val_metrics_custom['f1']:.4f}")
    Validation loss: 0.4215, F1: 0.7945
    """

    model.eval()
    running_loss = 0.0
    all_outputs = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(val_loader, desc='Validation'):
            # Handle batch structure based on whether motion features are included
            if include_motion_features and len(batch) == 3:
                inputs, labels, motion_features = batch
                inputs = inputs.to(device)
                labels = labels.to(device)
                motion_features = motion_features.to(device)
                outputs = model(inputs, motion_features)
            else:
                inputs, labels = batch
                inputs = inputs.to(device)
                labels = labels.to(device)
                outputs = model(inputs)

            loss = criterion(outputs.squeeze(), labels)

            running_loss += loss.item()
            all_outputs.extend(outputs.squeeze().cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    metrics = calculate_metrics(torch.tensor(all_outputs), torch.tensor(all_labels), threshold)
    metrics['loss'] = running_loss / len(val_loader)

    return metrics

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=NUM_EPOCHS,
                include_motion_features=True, experiment_name="default",
                patience=EARLY_STOP_PATIENCE, val_threshold=0.5):
    """
    Train model with comprehensive tracking and early stopping.

    Parameters:
    -----------
    model : torch.nn.Module
        Model to train
    train_loader : torch.utils.data.DataLoader
        DataLoader for training data
    val_loader : torch.utils.data.DataLoader
        DataLoader for validation data
    criterion : torch.nn.Module
        Loss function
    optimizer : torch.optim.Optimizer
        Optimizer
    num_epochs : int, default=NUM_EPOCHS
        Maximum number of epochs to train
    include_motion_features : bool, default=True
        Whether batches include motion features
    experiment_name : str, default="default"
        Name for the experiment (used for logging)
    patience : int, default=EARLY_STOP_PATIENCE
        Early stopping patience
    val_threshold : float, default=0.5
        Validation classification threshold

    Returns:
    --------
    tuple
        (model, history) - Trained model and training history dictionary

    Example:
    --------
    >>> # Setup model, criterion, optimizer
    >>> model = OptimizedPageFlipNet(include_motion_features=True)
    >>> criterion = nn.BCELoss()
    >>> optimizer = optim.Adam(model.parameters(), lr=0.001)
    >>>
    >>> # Train model
    >>> model, history = train_model(
    ...     model,
    ...     train_loader,
    ...     val_loader,
    ...     criterion,
    ...     optimizer,
    ...     num_epochs=10,
    ...     experiment_name="motion_features_model"
    ... )
    >>>
    >>> # Plot training history
    >>> plot_training_history(history, "Motion Features Model")
    """

    print(f"\n Starting training for experiment: {experiment_name}")

    early_stopping = EarlyStopping(patience=patience, restore_best_weights=True)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2, verbose=True)

    history = {
        'train_loss': [], 'train_metrics': [],
        'val_loss': [], 'val_metrics': [],
        'learning_rates': []
    }

    for epoch in range(num_epochs):
        print(f'\nEpoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        # Record learning rate
        current_lr = optimizer.param_groups[0]['lr']
        history['learning_rates'].append(current_lr)

        # Train and validate
        train_metrics = train_epoch(
            model, train_loader, criterion, optimizer,
            include_motion_features=include_motion_features
        )

        val_metrics = validate_epoch(
            model, val_loader, criterion,
            include_motion_features=include_motion_features,
            threshold=val_threshold
        )

        # Print detailed metrics
        print(f'Training - Loss: {train_metrics["loss"]:.4f}, '
              f'Acc: {train_metrics["accuracy"]:.4f}, '
              f'F1: {train_metrics["f1"]:.4f}, '
              f'Prec: {train_metrics["precision"]:.4f}, '
              f'Rec: {train_metrics["recall"]:.4f}')

        print(f'Validation - Loss: {val_metrics["loss"]:.4f}, '
              f'Acc: {val_metrics["accuracy"]:.4f}, '
              f'F1: {val_metrics["f1"]:.4f}, '
              f'Prec: {val_metrics["precision"]:.4f}, '
              f'Rec: {val_metrics["recall"]:.4f}')

        # Update learning rate based on validation loss
        scheduler.step(val_metrics['loss'])

        # Save model if it's the best so far
        model_improved = early_stopping(val_metrics['loss'], model)
        if model_improved:
            model_save_path = f'best_model_{experiment_name}.pth'
            torch.save(model.state_dict(), model_save_path)
            print(f" Model improved - saved checkpoint")

        # Update history
        history['train_loss'].append(train_metrics['loss'])
        history['train_metrics'].append(train_metrics)
        history['val_loss'].append(val_metrics['loss'])
        history['val_metrics'].append(val_metrics)

        # Check for early stopping
        if early_stopping.early_stop:
            print(f"⚠ Early stopping triggered after epoch {epoch+1}")
            if early_stopping.restore_model(model):
                print("Restored best model state")
            break

    return model, history

def print_model_size(model):
    """
    Calculate and print detailed model size information.

    Parameters:
    -----------
    model : torch.nn.Module
        Model to analyze

    Returns:
    --------
    tuple
        (num_params, size_all_mb) - Number of parameters and model size in MB

    Example:
    --------
    >>> model = OptimizedPageFlipNet(include_motion_features=True)
    >>> num_params, size_mb = print_model_size(model)
    >>> print(f"Model has {num_params:,} parameters and is {size_mb:.2f} MB in size")
    Model has 1,274,305 parameters and is 4.86 MB in size
    """

    param_size = 0
    num_params = 0

    # Calculate sizes by layer
    layer_sizes = {}
    for name, param in model.named_parameters():
        param_size += param.nelement() * param.element_size()
        num_params += param.nelement()

        # Extract layer name
        layer_name = name.split('.')[0]
        if layer_name not in layer_sizes:
            layer_sizes[layer_name] = 0
        layer_sizes[layer_name] += param.nelement() * param.element_size()

    # Calculate buffer size
    buffer_size = 0
    for buffer in model.buffers():
        buffer_size += buffer.nelement() * buffer.element_size()

    size_all_mb = (param_size + buffer_size) / 1024**2

    print(f'\n Model Information:')
    print(f'Total Parameters: {num_params:,}')
    print(f'Model Size: {size_all_mb:.3f} MB')

    # Print size by layer
    print("\nSize by component:")
    for layer, size in sorted(layer_sizes.items(), key=lambda x: x[1], reverse=True):
        size_mb = size / 1024**2
        print(f"  - {layer}: {size_mb:.3f} MB ({size_mb/size_all_mb*100:.1f}%)")

    return num_params, size_all_mb

def plot_confusion_matrix(y_true, y_pred, title="Confusion Matrix"):
    """
    Plot detailed confusion matrix with metrics.

    Parameters:
    -----------
    y_true : array-like
        Ground truth labels
    y_pred : array-like
        Predicted labels
    title : str, default="Confusion Matrix"
        Title for the plot

    Returns:
    --------
    tuple
        (cm, accuracy, precision, recall, specificity, f1) - Confusion matrix and metrics

    Example:
    --------
    >>> # Evaluate model on test set
    >>> with torch.no_grad():
    ...     predictions = []
    ...     labels = []
    ...     for batch in test_loader:
    ...         inputs, batch_labels = batch[:2]  # Handle different batch structures
    ...         outputs = model(inputs)
    ...         predictions.extend((outputs.squeeze() > 0.5).cpu().numpy())
    ...         labels.extend(batch_labels.cpu().numpy())
    >>>
    >>> # Plot confusion matrix
    >>> cm, acc, prec, rec, spec, f1 = plot_confusion_matrix(
    ...     labels,
    ...     predictions,
    ...     title="Test Set Confusion Matrix"
    ... )
    >>> print(f"Test F1 Score: {f1:.4f}")
    Test F1 Score: 0.8734
    """

    # Calculate confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Calculate metrics from confusion matrix
    tn, fp, fn, tp = cm.ravel()
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    # Create figure
    plt.figure(figsize=(10, 8))

    # Plot heatmap
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{title}\nAccuracy: {accuracy:.4f}', fontsize=14)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks([0.5, 1.5], ['Not Flip', 'Flip'])
    plt.yticks([0.5, 1.5], ['Not Flip', 'Flip'])

    # Add metrics text box
    metrics_text = f"Precision: {precision:.4f}\nRecall: {recall:.4f}\n" \
                  f"Specificity: {specificity:.4f}\nF1-Score: {f1:.4f}"
    plt.figtext(0.15, 0.01, metrics_text, fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

    plt.tight_layout()
    plt.show()

    return cm, accuracy, precision, recall, specificity, f1

def plot_training_history(history, experiment_name="default"):
    """
    Plot comprehensive training metrics with proper labeling.

    Parameters:
    -----------
    history : dict
        Training history dictionary from train_model
    experiment_name : str, default="default"
        Name for the experiment (used in plot title)

    Returns:
    --------
    None
        Displays and saves plots

    Example:
    --------
    >>> # After training model
    >>> model, history = train_model(...)
    >>>
    >>> # Plot training history
    >>> plot_training_history(history, "Page Flip Detection Model")
    """

    plt.figure(figsize=(20, 12))

    # Plot loss
    plt.subplot(2, 3, 1)
    plt.plot(history['train_loss'], label='Train Loss', marker='o')
    plt.plot(history['val_loss'], label='Val Loss', marker='x')
    plt.title('Loss over Epochs', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()

    # Plot accuracy
    plt.subplot(2, 3, 2)
    train_acc = [m['accuracy'] for m in history['train_metrics']]
    val_acc = [m['accuracy'] for m in history['val_metrics']]
    plt.plot(train_acc, label='Train Accuracy', marker='o')
    plt.plot(val_acc, label='Val Accuracy', marker='x')
    plt.title('Accuracy over Epochs', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()

    # Plot F1 Score
    plt.subplot(2, 3, 3)
    f1_train = [m['f1'] for m in history['train_metrics']]
    f1_val = [m['f1'] for m in history['val_metrics']]
    plt.plot(f1_train, label='Train F1', marker='o')
    plt.plot(f1_val, label='Val F1', marker='x')
    plt.title('F1 Score over Epochs', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('F1 Score')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()

    # Plot Precision
    plt.subplot(2, 3, 4)
    prec_train = [m['precision'] for m in history['train_metrics']]
    prec_val = [m['precision'] for m in history['val_metrics']]
    plt.plot(prec_train, label='Train Precision', marker='o')
    plt.plot(prec_val, label='Val Precision', marker='x')
    plt.title('Precision over Epochs', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Precision')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()

    # Plot Recall
    plt.subplot(2, 3, 5)
    rec_train = [m['recall'] for m in history['train_metrics']]
    rec_val = [m['recall'] for m in history['val_metrics']]
    plt.plot(rec_train, label='Train Recall', marker='o')
    plt.plot(rec_val, label='Val Recall', marker='x')
    plt.title('Recall over Epochs', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Recall')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()

    # Plot Learning Rate
    plt.subplot(2, 3, 6)
    plt.plot(history['learning_rates'], marker='s')
    plt.title('Learning Rate Schedule', fontsize=14)
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.yscale('log')

    plt.suptitle(f'Training Metrics - {experiment_name}', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Make room for suptitle
    plt.savefig(f'training_history_{experiment_name}.png')
    plt.show()

def find_optimal_threshold(model, val_loader, include_motion_features=True):
    """
    Find the optimal threshold that maximizes F1 score.

    Parameters:
    -----------
    model : torch.nn.Module
        Trained model
    val_loader : torch.utils.data.DataLoader
        DataLoader for validation data
    include_motion_features : bool, default=True
        Whether batches include motion features

    Returns:
    --------
    float
        Optimal threshold value

    Example:
    --------
    >>> # Find optimal threshold after training
    >>> optimal_threshold = find_optimal_threshold(model, val_loader)
    >>> print(f"Optimal classification threshold: {optimal_threshold:.2f}")
    Optimal classification threshold: 0.42
    >>>
    >>> # Use optimal threshold for evaluation
    >>> test_metrics = test_with_threshold(
    ...     model,
    ...     test_loader,
    ...     threshold=optimal_threshold
    ... )
    """

    print("\n Finding optimal classification threshold...")

    model.eval()
    all_outputs = []
    all_labels = []

    # Collect all predictions
    with torch.no_grad():
        for batch in tqdm(val_loader, desc='Collecting predictions'):
            if include_motion_features and len(batch) == 3:
                inputs, labels, motion_features = batch
                inputs = inputs.to(device)
                labels = labels.to(device)
                motion_features = motion_features.to(device)
                outputs = model(inputs, motion_features)
            else:
                inputs, labels = batch
                inputs = inputs.to(device)
                labels = labels.to(device)
                outputs = model(inputs)

            all_outputs.extend(outputs.squeeze().cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Convert to numpy arrays
    outputs_array = np.array(all_outputs)
    labels_array = np.array(all_labels)

    # Test different thresholds
    thresholds = np.linspace(0.1, 0.9, 17)  # Try thresholds from 0.1 to 0.9
    results = []

    for threshold in thresholds:
        predictions = (outputs_array > threshold).astype(int)

        # Calculate metrics
        precision = precision_score(labels_array, predictions, zero_division=0)
        recall = recall_score(labels_array, predictions)
        f1 = f1_score(labels_array, predictions)

        results.append({
            'threshold': threshold,
            'precision': precision,
            'recall': recall,
            'f1': f1
        })

        print(f"Threshold {threshold:.2f}: F1={f1:.4f}, Precision={precision:.4f}, Recall={recall:.4f}")

    # Find best threshold by F1 score
    best_result = max(results, key=lambda x: x['f1'])

    print(f"\n Optimal threshold: {best_result['threshold']:.2f} with F1={best_result['f1']:.4f}")
    print(f"    Precision: {best_result['precision']:.4f}, Recall: {best_result['recall']:.4f}")

    # Plot results
    plt.figure(figsize=(12, 6))

    plt.plot([r['threshold'] for r in results], [r['precision'] for r in results],
             marker='o', label='Precision')
    plt.plot([r['threshold'] for r in results], [r['recall'] for r in results],
             marker='s', label='Recall')
    plt.plot([r['threshold'] for r in results], [r['f1'] for r in results],
             marker='^', label='F1 Score')

    plt.axvline(x=best_result['threshold'], color='r', linestyle='--',
                label=f"Best threshold: {best_result['threshold']:.2f}")

    plt.title('Metrics vs. Threshold')
    plt.xlabel('Threshold')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return best_result['threshold']

def test_with_threshold(model, test_loader, threshold=0.5, include_motion_features=True):
    """
    Evaluate model on test set with specific threshold.

    Parameters:
    -----------
    model : torch.nn.Module
        Trained model
    test_loader : torch.utils.data.DataLoader
        DataLoader for test data
    threshold : float, default=0.5
        Classification threshold
    include_motion_features : bool, default=True
        Whether batches include motion features

    Returns:
    --------
    dict
        Dictionary containing test metrics

    Example:
    --------
    >>> # Evaluate model with default threshold
    >>> default_metrics = test_with_threshold(model, test_loader)
    >>> print(f"Default threshold metrics - F1: {default_metrics['f1']:.4f}")
    Default threshold metrics - F1: 0.8532
    >>>
    >>> # Evaluate with optimal threshold
    >>> optimal_threshold = find_optimal_threshold(model, val_loader)
    >>> optimal_metrics = test_with_threshold(model, test_loader, threshold=optimal_threshold)
    >>> print(f"Optimal threshold metrics - F1: {optimal_metrics['f1']:.4f}")
    Optimal threshold metrics - F1: 0.8765
    """

    model.eval()
    all_outputs = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(test_loader, desc=f'Testing (threshold={threshold:.2f})'):
            if include_motion_features and len(batch) == 3:
                inputs, labels, motion_features = batch
                inputs = inputs.to(device)
                labels = labels.to(device)
                motion_features = motion_features.to(device)
                outputs = model(inputs, motion_features)
            else:
                inputs, labels = batch
                inputs = inputs.to(device)
                labels = labels.to(device)
                outputs = model(inputs)

            all_outputs.extend(outputs.squeeze().cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Convert to numpy arrays
    outputs_array = np.array(all_outputs)
    labels_array = np.array(all_labels)

    # Apply threshold
    predictions = (outputs_array > threshold).astype(int)

    # Calculate metrics
    accuracy = (predictions == labels_array).mean()
    precision = precision_score(labels_array, predictions, zero_division=0)
    recall = recall_score(labels_array, predictions)
    f1 = f1_score(labels_array, predictions)

    # Print results
    print(f"\n Test Results (threshold={threshold:.2f}):")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    # Plot confusion matrix
    plot_confusion_matrix(labels_array, predictions, title=f"Test Set Confusion Matrix (threshold={threshold:.2f})")

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'predictions': predictions,
        'labels': labels_array,
        'outputs': outputs_array,
        'threshold': threshold
    }

def test_on_random_images(model, test_df, transform, num_samples=5,
                          include_motion_features=True, threshold=0.5):
    """
    Test model on random images with adjustable threshold and visualize results.

    Parameters:
    -----------
    model : torch.nn.Module
        Trained model
    test_df : pandas.DataFrame
        DataFrame containing test images
    transform : callable
        Transform to apply to images
    num_samples : int, default=5
        Number of random samples to test
    include_motion_features : bool, default=True
        Whether to include motion features
    threshold : float, default=0.5
        Classification threshold

    Returns:
    --------
    None
        Displays visualization of results

    Example:
    --------
    >>> # Test model on random images with optimal threshold
    >>> optimal_threshold = find_optimal_threshold(model, val_loader)
    >>> test_on_random_images(
    ...     model,
    ...     test_df,
    ...     val_transform,
    ...     num_samples=8,
    ...     threshold=optimal_threshold
    ... )
    """


    model.eval()

    print(f"\n Testing on random images (threshold={threshold:.2f})...")
    sample_df = test_df.sample(n=num_samples)

    plt.figure(figsize=(15, 4 * num_samples))

    for i, (_, row) in enumerate(sample_df.iterrows()):
        with Image.open(row['image_path']) as img:
            # Show original image
            plt.subplot(num_samples, 3, i*3 + 1)
            plt.imshow(img)
            plt.title(f"Original - {os.path.basename(row['image_path'])}")
            plt.axis('off')

            # Show preprocessed image
            img_processed = preprocess_image(img)
            plt.subplot(num_samples, 3, i*3 + 2)
            plt.imshow(img_processed)
            plt.title("Preprocessed")
            plt.axis('off')

            # Make prediction
            img_tensor = transform(img_processed).unsqueeze(0)
            # Move tensor to the same device as the model
            img_tensor = img_tensor.to(device)

            with torch.no_grad():
                if include_motion_features and set(row.index).intersection(['mean_motion', 'std_motion', 'max_motion']):
                    motion_features = torch.tensor([
                        row.get('mean_motion', 0),
                        row.get('std_motion', 0),
                        row.get('max_motion', 0)
                    ], dtype=torch.float32).unsqueeze(0)
                    # Move motion features to the same device as the model
                    motion_features = motion_features.to(device)
                    output = model(img_tensor, motion_features)
                else:
                    output = model(img_tensor)

                probability = output.item()
                prediction = 'Flip' if probability > threshold else 'Not Flip'
                true_label = 'Flip' if row['label'] == 1 else 'Not Flip'

                # Show prediction visualization
                plt.subplot(num_samples, 3, i*3 + 3)
                bar_heights = [1-probability, probability]
                bars = plt.bar(['Not Flip', 'Flip'], bar_heights, color=['skyblue', 'salmon'])
                plt.ylim(0, 1)
                plt.axhline(y=threshold, color='red', linestyle='--', label=f'Threshold ({threshold:.2f})')
                plt.title(f"Prediction: {prediction} ({probability:.2%})\nTrue: {true_label}")
                plt.legend()

                # Highlight correct/incorrect
                if (prediction == 'Flip' and true_label == 'Flip') or \
                   (prediction == 'Not Flip' and true_label == 'Not Flip'):
                    plt.gca().spines['bottom'].set_color('green')
                    plt.gca().spines['top'].set_color('green')
                    plt.gca().spines['left'].set_color('green')
                    plt.gca().spines['right'].set_color('green')
                    plt.gca().spines['bottom'].set_linewidth(2)
                    plt.gca().spines['top'].set_linewidth(2)
                    plt.gca().spines['left'].set_linewidth(2)
                    plt.gca().spines['right'].set_linewidth(2)
                else:
                    plt.gca().spines['bottom'].set_color('red')
                    plt.gca().spines['top'].set_color('red')
                    plt.gca().spines['left'].set_color('red')
                    plt.gca().spines['right'].set_color('red')
                    plt.gca().spines['bottom'].set_linewidth(2)
                    plt.gca().spines['top'].set_linewidth(2)
                    plt.gca().spines['left'].set_linewidth(2)
                    plt.gca().spines['right'].set_linewidth(2)

    plt.tight_layout()
    plt.show()


def test_on_custom_images(model, image_paths, transform, include_motion_features=False, threshold=0.5):
    """Test the model on user-provided custom images"""
    if not image_paths:
        print("No image paths provided!")
        return

    model.eval()

    print(f"\n Testing on custom images (threshold={threshold:.2f})...")
    plt.figure(figsize=(15, 5 * len(image_paths)))

    for i, img_path in enumerate(image_paths):
        if not os.path.exists(img_path):
            print(f"Image not found: {img_path}")
            continue

        with Image.open(img_path) as img:
            # Show original image
            plt.subplot(len(image_paths), 3, i*3 + 1)
            plt.imshow(img)
            plt.title(f"Original - {os.path.basename(img_path)}")
            plt.axis('off')

            # Show preprocessed image
            img_processed = preprocess_image(img)
            plt.subplot(len(image_paths), 3, i*3 + 2)
            plt.imshow(img_processed)
            plt.title("Preprocessed")
            plt.axis('off')

            # Make prediction
            img_tensor = transform(img_processed).unsqueeze(0)
            # Add this line to move tensor to the correct device
            img_tensor = img_tensor.to(device)

            with torch.no_grad():
                output = model(img_tensor)
                probability = output.item()
                prediction = 'Flip' if probability > threshold else 'Not Flip'

                # Show prediction visualization
                plt.subplot(len(image_paths), 3, i*3 + 3)
                bar_heights = [1-probability, probability]
                bars = plt.bar(['Not Flip', 'Flip'], bar_heights, color=['skyblue', 'salmon'])
                plt.ylim(0, 1)
                plt.axhline(y=threshold, color='red', linestyle='--', label=f'Threshold ({threshold:.2f})')
                plt.title(f"Prediction: {prediction} ({probability:.2%})")
                plt.legend()

    plt.tight_layout()
    plt.show()

    print(" Custom image testing complete")

def save_model_with_metadata(model, optimizer, history, test_metrics, training_time,
                            config, experiment_name="final"):
    """
    Save model with comprehensive metadata for future reference.

    Parameters:
    -----------
    model : torch.nn.Module
        Trained model
    optimizer : torch.optim.Optimizer
        Optimizer
    history : dict
        Training history
    test_metrics : dict
        Test metrics
    training_time : float
        Training time in seconds
    config : dict
        Model and training configuration
    experiment_name : str, default="final"
        Name for the experiment

    Returns:
    --------
    str
        Path to saved model

    Example:
    --------
    >>> # After training and evaluation
    >>> config = {
    ...     'image_size': 96,
    ...     'kernel_sizes': [3, 5, 3, 3],
    ...     'use_motion_features': True,
    ...     'optimal_threshold': 0.42
    ... }
    >>>
    >>> # Save model with metadata
    >>> save_path = save_model_with_metadata(
    ...     model,
    ...     optimizer,
    ...     history,
    ...     test_metrics,
    ...     training_time,
    ...     config,
    ...     experiment_name="motion_features_model"
    ... )
    >>> print(f"Model saved to {save_path}")
    Model saved to motion_features_model_model.pth
    """

    save_path = f'{experiment_name}_model.pth'

    # Create directory if it doesn't exist
    os.makedirs(os.path.dirname(save_path) if os.path.dirname(save_path) else '.', exist_ok=True)

    # Create timestamp
    timestamp = time.strftime("%Y%m%d-%H%M%S")

    # Save model with metadata
    torch.save({
        'timestamp': timestamp,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'history': history,
        'test_metrics': test_metrics,
        'training_time': training_time,
        'config': config,
        'experiment_name': experiment_name
    }, save_path)

    print(f" Model and metadata saved to {save_path}")

    # Save a summary text file with key metrics
    summary_path = f'{experiment_name}_summary.txt'
    with open(summary_path, 'w') as f:
        f.write(f"Experiment: {experiment_name}\n")
        f.write(f"Timestamp: {timestamp}\n")
        f.write(f"Training time: {training_time/60:.2f} minutes\n\n")

        f.write("Test Metrics:\n")
        for key, value in test_metrics.items():
            if key not in ['predictions', 'labels', 'outputs']:
                f.write(f"  {key}: {value:.4f}\n")

        f.write("\nConfiguration:\n")
        for key, value in config.items():
            f.write(f"  {key}: {value}\n")

    print(f" Summary saved to {summary_path}")

    return save_path


# GradCAM implementation for visualization
import cv2

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.feature_maps = None
        self.gradient = None

        # Register hooks to capture activations and gradients
        self.hook_handle = target_layer.register_forward_hook(self.save_features)
        self.hook_backward = target_layer.register_backward_hook(self.save_gradient)

    def save_features(self, module, input, output):
        self.feature_maps = output.detach()

    def save_gradient(self, module, grad_input, grad_output):
        self.gradient = grad_output[0].detach()

    def __call__(self, input_image, target_class=None):
        # Forward pass
        self.model.eval()
        output = self.model(input_image)

        # If target class is None, use the predicted class
        if target_class is None:
            target_class = 0  # We only have one output (sigmoid)

        # Clear any previous gradients
        self.model.zero_grad()

        # Backpropagate to calculate gradients
        output[0].backward()

        # Calculate weights based on global average pooling
        weights = torch.mean(self.gradient, dim=(2, 3))

        # Generate class activation map
        cam = torch.zeros(self.feature_maps.shape[2:], dtype=torch.float32)

        # Multiply each feature map with its corresponding weight
        for i, w in enumerate(weights[0]):
            cam += w * self.feature_maps[0, i]

        # Apply ReLU to consider only positive influence
        cam = torch.max(cam, torch.tensor(0.))

        # Normalize for visualization
        cam = cam - torch.min(cam)
        cam = cam / torch.max(cam) if torch.max(cam) > 0 else cam

        # Clean up hooks after use
        self.hook_handle.remove()
        self.hook_backward.remove()

        return cam.cpu().numpy()

def visualize_model_attention(model, img_tensor, original_img):
    """
    Visualize which parts of the image the model is focusing on using GradCAM.
    """
    # Ensure model is in evaluation mode
    model.eval()

    # Get the target layer for GradCAM (use an earlier convolutional layer)
    target_layer = model.features_block2[-3]  # Try using an earlier layer

    # Create a simplified forward function to get only convolutional output
    def get_conv_output(x):
        x = model.features_block1(x)
        x = model.features_block2(x)
        return x

    # Forward pass to get the feature maps
    with torch.no_grad():
        img_tensor_batch = img_tensor.unsqueeze(0).to(device)
        feature_maps = get_conv_output(img_tensor_batch)

    # Manually create heatmap by summing across channels
    # This is a simplified approach that doesn't use gradients
    feature_activation = feature_maps.sum(dim=1).squeeze().cpu().numpy()

    # Normalize for visualization
    feature_activation = feature_activation - feature_activation.min()
    feature_activation = feature_activation / feature_activation.max()

    # Resize to match original image
    original_img_np = np.array(original_img)
    heatmap_resized = cv2.resize(feature_activation, (original_img_np.shape[1], original_img_np.shape[0]))

    # Convert heatmap to RGB for visualization
    heatmap_colored = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
    heatmap_colored = cv2.cvtColor(heatmap_colored, cv2.COLOR_BGR2RGB)

    # Overlay heatmap on original image
    original_img_float = original_img_np.astype(float) / 255
    overlay = heatmap_colored.astype(float) / 255
    combined = original_img_float * 0.7 + overlay * 0.3

    # Ensure the values are in the valid range
    combined = np.clip(combined, 0, 1)

    return combined

def test_custom_images_simple(model, image_paths, transform, threshold=0.15):
    """
    Test the model on custom images with simplified attention visualization.
    """
    model.eval()

    plt.figure(figsize=(15, 4 * len(image_paths)))

    for i, img_path in enumerate(image_paths):
        try:
            with Image.open(img_path) as img:
                # Show original image
                plt.subplot(len(image_paths), 3, i*3 + 1)
                plt.imshow(img)
                plt.title(f"Original - {os.path.basename(img_path)}")
                plt.axis('off')

                # Preprocess image for model
                img_processed = preprocess_image(img)
                plt.subplot(len(image_paths), 3, i*3 + 2)
                plt.imshow(img_processed)
                plt.title(f"Preprocessed")
                plt.axis('off')

                # Convert to tensor for model
                img_tensor = transform(img_processed)
                # Add this line to move tensor to the correct device
                img_tensor = img_tensor.to(device)

                # Generate attention map without model prediction
                visualization = visualize_model_attention(model, img_tensor, img)

                # Show attention map
                plt.subplot(len(image_paths), 3, i*3 + 3)
                plt.imshow(visualization)
                plt.title(f"Model Attention")
                plt.axis('off')

        except Exception as e:
            print(f"Error processing {img_path}: {str(e)}")

    plt.tight_layout()
    plt.show()




if __name__ == "__main__":
    try:
        print(" Starting optimized page flip detection training...")
        start_total_time = time.time()

        # Create and analyze dataset
        df = create_dataset_df(base_path)
        analyze_dataset(df)

        # Calculate motion features if needed
        if USE_MOTION_FEATURES:
            df = calculate_optimized_motion_features(df, use_cache=True)

        # Visualize preprocessing options
        visualize_preprocessing(df, num_samples=2)

        # Get train and test split from existing structure
        train_df = df[df['split'] == 'training'].copy()
        test_df = df[df['split'] == 'testing'].copy()

        # Create validation split from training data with stratification
        train_df, val_df = train_test_split(
            train_df,
            test_size=0.2,
            stratify=train_df['label'],
            random_state=SEED
        )

        # Print split sizes
        print("\n Data Split Summary:")
        print(f"Training: {len(train_df)} images")
        print(f"Validation: {len(val_df)} images")
        print(f"Testing: {len(test_df)} images")

        # Get transforms (with or without augmentation)
        train_transform, val_transform = get_transforms(use_augmentation=True)

        # Create datasets with optimized configuration
        train_dataset = PageFlipDataset(
            train_df,
            transform=train_transform,
            include_motion_features=USE_MOTION_FEATURES,
            preprocessing_level='basic',  # Basic preprocessing for faster training
            augment=True  # Subtle augmentation helps with generalization
        )

        val_dataset = PageFlipDataset(
            val_df,
            transform=val_transform,
            include_motion_features=USE_MOTION_FEATURES,
            preprocessing_level='basic',
            augment=False  # No augmentation for validation
        )

        test_dataset = PageFlipDataset(
            test_df,
            transform=val_transform,
            include_motion_features=USE_MOTION_FEATURES,
            preprocessing_level='basic',
            augment=False  # No augmentation for testing
        )

        # Create optimized dataloaders with larger batch size
        train_loader = DataLoader(
            train_dataset,
            batch_size=BATCH_SIZE,
            shuffle=True,
            num_workers=WORKERS,
            pin_memory=True,
            persistent_workers=WORKERS > 0
        )

        val_loader = DataLoader(
            val_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=WORKERS,
            pin_memory=True,
            persistent_workers=WORKERS > 0
        )

        test_loader = DataLoader(
            test_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=WORKERS,
            pin_memory=True
        )

        # Create and print model
        # Create and print model
        model = OptimizedPageFlipNet(
            include_motion_features=USE_MOTION_FEATURES,
            kernel_sizes=[3, 5, 3, 3],  # Varied kernel sizes as recommended
            dropout_rate=0.3  # Moderate dropout to prevent overfitting
        )
        model = model.to(device)  # Move model to GPU if available
        print_model_size(model)

        # Setup training with suitable hyperparameters
        criterion = nn.BCELoss()
        optimizer = optim.Adam(
            model.parameters(),
            lr=0.001,
            weight_decay=0.0001 if USE_REGULARIZATION else 0  # L2 regularization
        )

        # Train model with timing
        print("\n Starting optimized training...")
        start_time = time.time()

        model, history = train_model(
            model,
            train_loader,
            val_loader,
            criterion,
            optimizer,
            num_epochs=NUM_EPOCHS,
            include_motion_features=USE_MOTION_FEATURES,
            experiment_name="optimized",
            patience=EARLY_STOP_PATIENCE
        )

        training_time = time.time() - start_time
        print(f"\n Training completed in {training_time/60:.2f} minutes")

        # Plot training history
        plot_training_history(history, "Optimized Model")

        # Find optimal threshold based on validation set
        optimal_threshold = find_optimal_threshold(model, val_loader, include_motion_features=USE_MOTION_FEATURES)

        # Evaluate on test set with optimal threshold
        print("\n Evaluating on test set with optimal threshold...")
        test_metrics = test_with_threshold(
            model,
            test_loader,
            threshold=optimal_threshold,
            include_motion_features=USE_MOTION_FEATURES
        )

        # Save model with comprehensive information
        config = {
            'image_size': IMAGE_SIZE,
            'kernel_sizes': [3, 5, 3, 3],
            'preprocessing_level': 'basic',
            'use_augmentation': True,
            'use_motion_features': USE_MOTION_FEATURES,
            'use_regularization': USE_REGULARIZATION,
            'prevent_overfitting': PREVENT_OVERFITTING,
            'dropout_rate': 0.3,
            'optimal_threshold': optimal_threshold,
            'batch_size': BATCH_SIZE
        }

        save_model_with_metadata(
            model,
            optimizer,
            history,
            test_metrics,
            training_time,
            config,
            experiment_name="optimized"
        )

        # Test on random images from test set with optimal threshold
        test_on_random_images(
            model,
            test_df,
            val_transform,
            num_samples=5,
            include_motion_features=USE_MOTION_FEATURES,
            threshold=optimal_threshold
        )

        # Test on custom images if available
        custom_image_dir = os.path.join(base_path, 'custom_images')
        if os.path.exists(custom_image_dir):
            custom_images = [os.path.join(custom_image_dir, f)
                           for f in os.listdir(custom_image_dir)
                           if f.endswith(('.jpg', '.jpeg', '.png'))]

            if custom_images:
                test_on_custom_images(
                    model,
                    custom_images,
                    val_transform,
                    include_motion_features=False,  # Usually no motion features for custom images
                    threshold=optimal_threshold
                )

        # Print final summary
        total_time = time.time() - start_total_time

        print("\n Final Optimization Summary:")
        print(f"Total execution time: {total_time/60:.2f} minutes")
        print(f"Training time only: {training_time/60:.2f} minutes")
        print(f"Best F1 Score: {test_metrics['f1']:.4f}")
        print(f"Final Precision: {test_metrics['precision']:.4f}")
        print(f"Final Recall: {test_metrics['recall']:.4f}")
        print(f"Optimal threshold: {optimal_threshold:.2f}")

        # Print key improvements
        print("\n Key Improvements:")
        print("• Motion feature extraction optimized")
        print("• Model size reduced while maintaining performance")
        print("• Training/validation stability improved with regularization")
        print("• Recall enhanced with optimal threshold adjustment")
        print("• Total execution time significantly reduced")

        print("\n Optimized training and evaluation complete!")


        # After your model has been trained and evaluated
        print("\n Testing GradCAM on custom images...")

    # Make sure you're using the best model
        model.load_state_dict(torch.load('best_model_optimized.pth'))

        custom_images = [
            "/content/drive/MyDrive/IMG_20250303_135857_857.jpg",
            "/content/drive/MyDrive/IMG_20250303_135856_534.jpg",
            "/content/drive/MyDrive/IMG_20250303_135855_250.jpg"
        ]


        # Test with simplified visualization
        test_custom_images_simple(
            model,
            custom_images,
            val_transform
        )

        print("\n Optimized training and evaluation complete!")

    except Exception as e:
        print(f"\n Error occurred: {str(e)}")
        import traceback
        traceback.print_exc()
```
